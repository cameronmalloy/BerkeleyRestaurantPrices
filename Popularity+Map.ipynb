```python
from datascience import *
import math
import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plots
plots.style.use('fivethirtyeight')
```

```python
data = pd.read_csv('data.csv')
data
```

```python
#Calculates distance from UC Berkeley
latitude = np.array(data.iloc[:, 0])
longitude = np.array(data.iloc[:, 1])
longitude_abs = np.abs(longitude)
distance_from_ucb = (np.abs(latitude - 37.8716) + np.abs(longitude_abs - 122.2727)) ** 0.5
data['Distance From Berkeley'] = distance_from_ucb
```

```python
#Calculates most popular restaurants
rating = np.array(data.loc[:, 'Rating'])
review_count = np.array(data.loc[:, 'Review Count'])
#popularity = (rating * 2) + (review_count // 600) + rating * (review_count // 100)
popularity = 3 ** (rating * np.log10(review_count)) / 1000
data['Popularity'] = popularity
data_popularity_sorted = data.sort('Popularity', ascending=False)
data_popularity_sorted
```

```python
p_latitude = np.array(data_popularity_sorted.loc[:,'Latitude'])
p_longitude = np.array(data_popularity_sorted.loc[:,'Longitude'])
p_name = np.array(data_popularity_sorted.loc[:,'Restaurant Name'])
p_popularity = np.array(data_popularity_sorted.loc[:,'Popularity'])
p_prices = np.array(data_popularity_sorted.loc[:, 'Prices'])
map_tbl = Table().with_column(
    'lat', p_latitude,
    'lon', p_longitude,
    'label', p_name,
    'prices', p_prices,
    'Popularity', p_popularity
)
map_tbl_fin = map_tbl.where('Popularity', are.above(621))

color = make_array()
for i in np.arange(map_tbl_fin.num_rows):
    price = map_tbl_fin.column('prices').item(i)
    color_str = 'purple'
    if price == 1:
        color_str = 'green'
    elif price == 2:
        color_str = 'blue'
    elif price == 3:
        color_str = 'orange'
    elif price == 4:
        color_str = 'red'
    color = np.append(color, color_str)

popularity_fin = make_array()
for elm in map_tbl_fin.column('Popularity'):
    popularity_fin = np.append(popularity_fin, int(elm) * 30)
map_tbl_fin = map_tbl_fin.drop('Popularity').drop('prices').with_column('color', color, 'radius', popularity_fin)
Circle.map_table(map_tbl_fin)
```
